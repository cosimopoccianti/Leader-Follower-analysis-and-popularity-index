```python
def adfuller_test(series, signif=0.05, name='', verbose=False):
    """Perform ADFuller to test for Stationarity of given series and print report"""
    r = adfuller(series, autolag='AIC')
    output = {'test_statistic':round(r[0], 4), 'pvalue':round(r[1], 4), 'n_lags':round(r[2], 4), 'n_obs':r[3]}
    p_value = output['pvalue'] 
    def adjust(val, length= 6): return str(val).ljust(length)

    # Print Summary
    print(f'    Augmented Dickey-Fuller Test on "{name}"', "\n   ", '-'*47)
    print(f' Null Hypothesis: Data has unit root. Non-Stationary.')
    print(f' Significance Level    = {signif}')
    print(f' Test Statistic        = {output["test_statistic"]}')
    print(f' No. Lags Chosen       = {output["n_lags"]}')

    for key,val in r[4].items():
        print(f' Critical value {adjust(key)} = {round(val, 3)}')

    if p_value <= signif:
        print(f" => P-Value = {p_value}. Rejecting Null Hypothesis.")
        print(f" => Series is Stationary.")
    else:
        print(f" => P-Value = {p_value}. Weak evidence to reject the Null Hypothesis.")
        print(f" => Series is Non-Stationary.")    
```

The aim is to verify the presence of automated price systems and the presence of leaders and followers for November as well. The method used in the quarter analysis requires stationarity for all time series. 
We therefore tested this option with AD Fuller Test, and not all time series were stationary. So we tried a different approach for November, with the aim of identifying leaders and followers. 

```python
df=novembre.copy()

#absolute value of the price change
df['price_change'] = df['price_change'].abs()

##fill na with 0
df['price_change'] = df['price_change'].fillna(0) 
df["yes_no"] = np.where(df["price_change"] > 0, 1, 0)

#filter only for yes_no = 1 (yes means that there is a price change)
df = df[df['yes_no'] == 1]

df.sort_values(['comp_date'], inplace=True)
```

For each product, the difference (in terms of days) of each price change made by a seller was calculated

```python
df["comp_date"] = pd.to_datetime(df["comp_date"])
df["days_diff"]=df.groupby(['product_id'])["comp_date"].diff().dt.days
```

```python
#calculate the mean of the days_diff for each seller and product
df2=df.copy()
df2=df2.groupby(['product_id','seller_id'])['days_diff'].mean()

#discard the rows with nan values
df2=df2.dropna()

#convert the series to dataframe
df2=pd.DataFrame(df2)
```

The mean of the difference in terms of days when the seller changes the price of a speficic product 

```python
df2.head()
```

```python
#show df2 without base 10 exponential
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
#for each product_id sort the sellers by the mean of the days_diff
df2=df2.groupby(['product_id']).apply(lambda x: x.sort_values(['days_diff'], ascending=True))
df2
```

```python
#for each product_id calculate the difference betweene each seller
df2['diff'] = df2.groupby(['product_id'])['days_diff'].diff()
```

The 'diff' column shows the distance between the price changes of each seller. for example for product 100043, sellers 24 and 26 change their price on average every three days, so their price change frequency is similar. This may indicate a synchronicity in setting prices, thus the presentation of dynamic pricing.

# Leaders and Followers Analysis

```python
times_change=df.groupby(["product_id", "seller_id","comp_date"])["yes_no"].sum()
times_change=pd.DataFrame(times_change)

#remove levels
times_change=times_change.reset_index(level=[0,1,2])
```

'yes_no' column means that the product has undergone price changes 

```python
times_change
```

```python
#count times the price changed for each product_id and seller_id
count_change =times_change.groupby(["product_id", "seller_id"])["yes_no"].sum()
```

```python
#count_change to dataframe
count_change=pd.DataFrame(count_change)
count_change
```

```python
#map value yes_no for each product_id and seller_id to times_change and create new column
#the column represents the amount of times that the product changes the price
times_change['amount_change']=times_change.set_index(['product_id', 'seller_id']).index.map(count_change['yes_no'])
```

```python
#drop times_change
#times_change.drop("times_change", axis=1, inplace=True)
```

```python
#remove duplicate where product_id seller_id and amount_change are the same
#or each seller there are various repetitions in the dates, we took the least recent date (keep = 'first) because it indicates the first change price
times_change.drop_duplicates(subset=['product_id', 'seller_id', 'amount_change'],inplace=True, keep='first')
times_change
```

We have observed that, if automated pricing between two sellers is assumed, they will be inclined to follow the same trend; one imitates the other. So the number of times they change price is the same, which is why we only kept sellers who have an equal number of times they changed price.

```python
#remove row where amount_change is different for each product and seller_id
times_change=times_change[times_change.duplicated(subset=['product_id', 'amount_change'], keep=False)]
times_change.head(10)
```

```python
#for each product_id extract date that comes earlier which is the Leader date, and the date that comes later which is the Follower date
times_change['Leader_date'] = times_change.groupby(['product_id'])['comp_date'].transform('min')
times_change['Follower_date'] = times_change.groupby(['product_id'])['comp_date'].transform('max')


#if the Leader date is the same as the comp_date then the seller is the Leader
times_change['Lead/Follow'] = np.where(times_change['Leader_date'] == times_change['comp_date'], "Leader", "Follower")
```

```python
#for each seller_id count the number of times that the seller is the Leader
times_change['Leader_count'] = times_change.groupby(['seller_id'])['Lead/Follow'].transform('count')
```

```python
#number of times that each seller is the Leader
Leader_count=times_change.groupby(['seller_id'])['Leader_count'].max()
Leader_count=pd.DataFrame(Leader_count)
Leader_count
```

```python
#sort Leader_count by the number of times that each seller is the Leader
Leader_count=Leader_count.sort_values(['Leader_count'], ascending=False)
```
